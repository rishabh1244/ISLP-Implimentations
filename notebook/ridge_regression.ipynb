```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/housing_demo.csv")
np.set_printoptions(precision=2, suppress=True)

X = data.to_numpy()[:,:-1]
Y = data.to_numpy()[:,-1]

print("X : ",X)
print("Y : ",Y)

# dimention = x.size() * x[0].size()
rng = np.random.default_rng(seed=42)
beta_param = rng.random(size=(len(X[0]), len(X)))
print("BETA : ",beta_param)
```

```python
# impliment ridge regression 

L = 0.3;
I = np.eye((X.T @ X).shape[0])
B_RIDGE = np.linalg.inv(X.T @ X + L * I ) @ X.T * Y 
print(B_RIDGE)
```
