```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/housing_demo.csv")
np.set_printoptions(precision=2, suppress=True)

X = data.to_numpy()[:,1:-1]
Y = data.to_numpy()[:,-1]

print("X : ",X)
print("Y : ",Y)
```

```python
# impliment ridge regression 
L = 10010;
I = np.eye(X.shape[1])
L = 100000
B_RIDGE = np.linalg.inv(X.T @ X + L * I ) @ X.T @ Y 
print(B_RIDGE)
```

```python
# impliment linear regression
B_LIN = np.linalg.inv(X.T @ X ) @ X.T @ Y 
print(B_LIN)

```

```python
# ploting both of them 
# for 2d models 
plt.style.use("dark_background")
plt.scatter(X, Y, color="white")

plt.plot(X, X * B_RIDGE, color="red", label="RIDGE")
plt.plot(X, X * B_LIN, color="blue", label="RIDGE")

plt.xlabel("size_sqft")
plt.ylabel("price_usd")
plt.legend()
plt.show()




# linear regression minimisses RSS 
#  ridge regression minimises rss + L * Beta , thats means the Beta matrix is also minimised
#  making the coefficient minimum
# B1 = [2,3] rss = 100 
# B2 = [99,100] rss = 99
# linear regression picks b2 cuz low rss 
# but ridge picks b1 cuz low B (2,3) < (99,100) , which makes the data less sensitive to inputs , which is desirable
```
